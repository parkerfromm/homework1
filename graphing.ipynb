```python
import pandas as pd
```

```python
file = "output_g.csv"
df = pd.read_csv(file, header=None)
```

```python
df
```

```python
for i in range(1,9):
    df.iloc[i] = abs(df.iloc[i] -df.iloc[0])
```

```python
df
```

```python
import math as m
fwd= []
cen =[]
h =[0.1, 0.01, 0.001, 0.0001]
h = [ m.log(val) for val in h]

for i in range (1,9):
    avg = m.log(df.iloc[i].mean())
    if i%2 ==0:
        cen.append(avg)
    else:
       fwd.append(avg)
```

```python
fwd
```

```python
import matplotlib.pyplot as plt



plt.plot(fwd, h, label ="fwd")
plt.plot(cen, h, label ="cen")
plt.xlabel("h")
plt.ylabel("abs error")
plt.legend()
```

The slope of the graph of the log of the absolute errors shows that the error is increasing with the forward difference as h gets smaller but is decreasing with the central difference.

```python
import math as m
import numpy as np
fwdx= []
fwdy= []
fwdz= []
cenx =[]
ceny =[]
cenz =[]

h =np.log([0.1, 0.01, 0.001, 0.0001])

for i in range (1,9):
    avg1 = m.log(df.iloc[i][0:4].mean())
    avg2 = m.log2(df.iloc[i][4:8].mean())
    avg3 = m.log2(df.iloc[i][8:12].mean())

    if i%2 ==0:
        cenx.append(avg1)
        ceny.append(avg2)
        cenz.append(avg3)
    else:
       fwdx.append(avg1)
       fwdx.append(avg2)
       fwdx.append(avg3)
```

```python
plt.plot(fwdx, h, label ="fwdx")
plt.plot(fwdy, h, label ="fwdy")
plt.plot(fwdz, h, label ="fwdz")
plt.plot(cenx, h, label ="cenx")
plt.plot(ceny, h, label ="ceny")
plt.plot(cenz, h, label ="cenz")
plt.xlabel("h")
plt.ylabel("abs error")
plt.legend()
```
